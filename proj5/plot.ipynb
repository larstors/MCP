```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize
```

```python
energy = np.genfromtxt("energy.txt")
dens = np.genfromtxt("dens.txt")
```

```python
plt.plot(energy[::100, 0], (energy[::100, 1]))
plt.xlabel(r"it")
plt.ylabel(r"$E_T$")
plt.grid()

plt.savefig("A2_1.pdf", dpi=200)
```

```python
fig6, ax = plt.subplots(nrows=3, ncols=1)
plt.tight_layout()
ax[0].hist(dens[:, 0], bins=200, density=True, label="Electron 1")
ax[1].hist(dens[:, 1], bins=200, density=True, label="Electron 2")
ax[2].hist(dens[:, 2], bins=200, density=True)

ax[0].set_ylabel(r"$\rho(r_1)$")
ax[1].set_ylabel(r"$\rho(r_2)$")
ax[2].set_ylabel(r"$\rho(r_{12})$")
ax[0].set_xlabel(r"$r_1$")
ax[1].set_xlabel(r"$r_2$")
ax[2].set_xlabel(r"$r_{12}$")
ax[0].grid()
ax[1].grid()
ax[2].grid()
plt.savefig("A2_2.pdf", dpi=200, bbox_inches="tight")
```
