```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as opt
plt.rcParams.update({'font.size': 10})
```

## Velocity distribution

```python
vd = np.genfromtxt("vel_dist.txt", delimiter=" ")

fig, ax = plt.subplots(nrows=2, ncols=2)
plt.tight_layout()
size = 20

ax[0, 0].hist(vd[:, 0], bins=30, color="r", density=True)
ax[0, 1].hist(vd[:, 1], bins=30, color="g", density=True)
ax[1, 0].hist(vd[:, 2], bins=30, color="b", density=True)
ax[1, 1].hist(vd[:, 3], bins=30, color="m", density=True)

ax[0, 0].annotate(r"$v_x$", xy=(0,0), xytext=(.5, 2), fontsize=size)
ax[1, 0].annotate(r"$v_z$", xy=(0,0), xytext=(-.5, 2), fontsize=size)
ax[0, 1].annotate(r"$v_y$", xy=(0,0), xytext=(.5, 2), fontsize=size)
ax[1, 1].annotate(r"$v$", xy=(0,0), xytext=(.6, 3), fontsize=size)

ax[0, 0].set_xlabel(r"$v_x$")
ax[0, 0].set_ylabel(r"$P(v_x)$")

ax[0, 1].set_xlabel(r"$v_y$")
ax[0, 1].set_ylabel(r"$P(v_y)$")

ax[1, 0].set_xlabel(r"$v_z$")
ax[1, 0].set_ylabel(r"$P(v_z)$")

ax[1, 1].set_xlabel(r"$v$")
ax[1, 1].set_ylabel(r"$P(v)$")

ax[0, 0].grid()
ax[0, 1].grid()
ax[1, 0].grid()
ax[1, 1].grid()

print(vd.max())

plt.savefig("velocity_distribution.pdf", dpi=150, bbox_inches="tight")

# #################### WITH TABLE ##########################

vd = np.genfromtxt("vel_dist_tab.txt", delimiter=" ")

fig1, ax1 = plt.subplots(nrows=2, ncols=2)
plt.tight_layout()
size = 20

ax1[0, 0].hist(vd[:, 0], bins=30, color="r", density=True)
ax1[0, 1].hist(vd[:, 1], bins=30, color="g", density=True)
ax1[1, 0].hist(vd[:, 2], bins=30, color="b", density=True)
ax1[1, 1].hist(vd[:, 3], bins=30, color="m", density=True)

ax1[0, 0].annotate(r"$v_x$", xy=(0,0), xytext=(.5, 2), fontsize=size)
ax1[1, 0].annotate(r"$v_z$", xy=(0,0), xytext=(-.5, 2), fontsize=size)
ax1[0, 1].annotate(r"$v_y$", xy=(0,0), xytext=(.5, 2), fontsize=size)
ax1[1, 1].annotate(r"$v$", xy=(0,0), xytext=(.6, 3), fontsize=size)

ax1[0, 0].set_xlabel(r"$v_x$")
ax1[0, 0].set_ylabel(r"$P(v_x)$")

ax1[0, 1].set_xlabel(r"$v_y$")
ax1[0, 1].set_ylabel(r"$P(v_y)$")

ax1[1, 0].set_xlabel(r"$v_z$")
ax1[1, 0].set_ylabel(r"$P(v_z)$")

ax1[1, 1].set_xlabel(r"$v$")
ax1[1, 1].set_ylabel(r"$P(v)$")

ax1[0, 0].grid()
ax1[0, 1].grid()
ax1[1, 0].grid()
ax1[1, 1].grid()

print(vd[:, 3].max())

plt.savefig("velocity_distribution.pdf", dpi=150, bbox_inches="tight")
```

## MELTING FACTOR

```python
rk = np.genfromtxt("melting_factor.txt", delimiter=" ")
rk_tab = np.genfromtxt("melting_factor_tab.txt", delimiter=" ")



fig = plt.figure()
plt.plot(rk[:, 0], rk[:, 1], "-o", label="without table")
plt.plot(rk_tab[:, 0], rk_tab[:, 1], "r-x", label="with table")
plt.xlabel(r"$t^{*}$")
plt.ylabel(r"$\rho_k$")
plt.grid()
plt.legend()
plt.show()


print(3*864, rk[0, 1], rk_tab[0, 1])
```

## TEMPERATURE
