```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as opt
plt.rcParams.update({'font.size': 10})
```

## Velocity distribution

```python
vd = np.genfromtxt("vel_dist_a.txt", delimiter=" ")

fig, ax = plt.subplots(nrows=2, ncols=2)
plt.tight_layout()
size = 20


def gauss(x, mu, sig):
    return 1/np.sqrt(2*np.pi*sig**2)*np.exp(-1/(2*sig**2) * (x - mu)**2)

def maxwell(x, a):
    return np.sqrt(2/np.pi) * x**2 * np.exp(-x**2 / (2*a**2))/ a**3

x_range1 = []
datahist1, bins1 = np.histogram(vd[:, 0], 30, density="True")

for i in range(len(bins1)-1):
    x_range1.append((bins1[i+1] + bins1[i])/2)

# parameters for gaussian fit
par1 = opt.curve_fit(gauss, xdata=x_range1, ydata=datahist1, p0=(0, 1))[0]


x_range1 = []
datahist1, bins1 = np.histogram(vd[:, 1], 30, density="True")

for i in range(len(bins1)-1):
    x_range1.append((bins1[i+1] + bins1[i])/2)

# parameters for gaussian fit
par2 = opt.curve_fit(gauss, xdata=x_range1, ydata=datahist1, p0=(0, 1))[0]


x_range1 = []
datahist1, bins1 = np.histogram(vd[:, 2], 30, density="True")

for i in range(len(bins1)-1):
    x_range1.append((bins1[i+1] + bins1[i])/2)

# parameters for gaussian fit
par3 = opt.curve_fit(gauss, xdata=x_range1, ydata=datahist1, p0=(0, 1))[0]

x_range1 = []
datahist1, bins1 = np.histogram(vd[:, 3], 30, density="True")

for i in range(len(bins1)-1):
    x_range1.append((bins1[i+1] + bins1[i])/2)

# parameters for gaussian fit
par4 = opt.curve_fit(maxwell, xdata=x_range1, ydata=datahist1, p0=(1))[0]


ax[0, 0].hist(vd[:, 0], bins=30, color="r", density=True)
ax[0, 0].plot(np.linspace(vd[:, 0].min(), vd[:, 0].max(), 100), gauss(np.linspace(vd[:, 0].min(), vd[:, 0].max(), 100), par1[0], par1[1]))
ax[0, 1].hist(vd[:, 1], bins=30, color="g", density=True)
ax[0, 1].plot(np.linspace(vd[:, 1].min(), vd[:, 1].max(), 100), gauss(np.linspace(vd[:, 1].min(), vd[:, 1].max(), 100), par2[0], par2[1]), "r-")
ax[1, 0].hist(vd[:, 2], bins=30, color="b", density=True)
ax[1, 0].plot(np.linspace(vd[:, 2].min(), vd[:, 2].max(), 100), gauss(np.linspace(vd[:, 2].min(), vd[:, 2].max(), 100), par3[0], par3[1]), "r-")
ax[1, 1].hist(vd[:, 3], bins=30, color="m", density=True)
ax[1, 1].plot(np.linspace(vd[:, 3].min(), vd[:, 3].max(), 100), maxwell(np.linspace(vd[:, 3].min(), vd[:, 3].max(), 100), par4[0]), "-g")


# ax[0, 0].annotate(r"$v_x$", xy=(0,0), xytext=(.3, 2), fontsize=size)
# ax[1, 0].annotate(r"$v_z$", xy=(0,0), xytext=(-.3, 2), fontsize=size)
# ax[0, 1].annotate(r"$v_y$", xy=(0,0), xytext=(.3, 2), fontsize=size)
# ax[1, 1].annotate(r"$v$", xy=(0,0), xytext=(.4, 3), fontsize=size)

ax[0, 0].set_xlabel(r"$v_x$")
ax[0, 0].set_ylabel(r"$P(v_x)$")

ax[0, 1].set_xlabel(r"$v_y$")
ax[0, 1].set_ylabel(r"$P(v_y)$")

ax[1, 0].set_xlabel(r"$v_z$")
ax[1, 0].set_ylabel(r"$P(v_z)$")

ax[1, 1].set_xlabel(r"$v$")
ax[1, 1].set_ylabel(r"$P(v)$")

ax[0, 0].grid()
ax[0, 1].grid()
ax[1, 0].grid()
ax[1, 1].grid()

print(vd.max())

plt.savefig("velocity_distribution_a.pdf", dpi=150, bbox_inches="tight")

# #################### WITH TABLE ##########################

vd = np.genfromtxt("vel_dist_tab_16_a.txt", delimiter=" ")

x_range1 = []
datahist1, bins1 = np.histogram(vd[:, 0], 30, density="True")

for i in range(len(bins1)-1):
    x_range1.append((bins1[i+1] + bins1[i])/2)

# parameters for gaussian fit
par1 = opt.curve_fit(gauss, xdata=x_range1, ydata=datahist1, p0=(0, 1))[0]


x_range1 = []
datahist1, bins1 = np.histogram(vd[:, 1], 30, density="True")

for i in range(len(bins1)-1):
    x_range1.append((bins1[i+1] + bins1[i])/2)

# parameters for gaussian fit
par2 = opt.curve_fit(gauss, xdata=x_range1, ydata=datahist1, p0=(0, 1))[0]


x_range1 = []
datahist1, bins1 = np.histogram(vd[:, 2], 30, density="True")

for i in range(len(bins1)-1):
    x_range1.append((bins1[i+1] + bins1[i])/2)

# parameters for gaussian fit
par3 = opt.curve_fit(gauss, xdata=x_range1, ydata=datahist1, p0=(0, 1))[0]

x_range1 = []
datahist1, bins1 = np.histogram(vd[:, 3], 30, density="True")

for i in range(len(bins1)-1):
    x_range1.append((bins1[i+1] + bins1[i])/2)

# parameters for gaussian fit
par4 = opt.curve_fit(maxwell, xdata=x_range1, ydata=datahist1, p0=(1))[0]


fig1, ax1 = plt.subplots(nrows=2, ncols=2)
plt.tight_layout()
size = 20

ax1[0, 0].hist(vd[:, 0], bins=30, color="r", density=True)
ax1[0, 0].plot(np.linspace(vd[:, 0].min(), vd[:, 0].max(), 100), gauss(np.linspace(vd[:, 0].min(), vd[:, 0].max(), 100), par1[0], par1[1]))
ax1[0, 1].hist(vd[:, 1], bins=30, color="g", density=True)
ax1[0, 1].plot(np.linspace(vd[:, 1].min(), vd[:, 1].max(), 100), gauss(np.linspace(vd[:, 1].min(), vd[:, 1].max(), 100), par2[0], par2[1]), "r-")
ax1[1, 0].hist(vd[:, 2], bins=30, color="b", density=True)
ax1[1, 0].plot(np.linspace(vd[:, 2].min(), vd[:, 2].max(), 100), gauss(np.linspace(vd[:, 2].min(), vd[:, 2].max(), 100), par3[0], par3[1]), "r-")
ax1[1, 1].hist(vd[:, 3], bins=30, color="m", density=True)
ax1[1, 1].plot(np.linspace(vd[:, 3].min(), vd[:, 3].max(), 100), maxwell(np.linspace(vd[:, 3].min(), vd[:, 3].max(), 100), par4[0]), "-g")



# ax1[0, 0].annotate(r"$v_x$", xy=(0,0), xytext=(.3, 2), fontsize=size)
# ax1[1, 0].annotate(r"$v_z$", xy=(0,0), xytext=(-.3, 2), fontsize=size)
# ax1[0, 1].annotate(r"$v_y$", xy=(0,0), xytext=(.3, 2), fontsize=size)
# ax1[1, 1].annotate(r"$v$", xy=(0,0), xytext=(.4, 3), fontsize=size)

ax1[0, 0].set_xlabel(r"$v_x$")
ax1[0, 0].set_ylabel(r"$P(v_x)$")

ax1[0, 1].set_xlabel(r"$v_y$")
ax1[0, 1].set_ylabel(r"$P(v_y)$")

ax1[1, 0].set_xlabel(r"$v_z$")
ax1[1, 0].set_ylabel(r"$P(v_z)$")

ax1[1, 1].set_xlabel(r"$v$")
ax1[1, 1].set_ylabel(r"$P(v)$")

ax1[0, 0].grid()
ax1[0, 1].grid()
ax1[1, 0].grid()
ax1[1, 1].grid()

print(vd[:, 3].max())

plt.savefig("velocity_distribution_tab_a.pdf", dpi=150, bbox_inches="tight")
```

```python
m1 = np.genfromtxt("msd_e_05.txt", delimiter=" ")
m2 = np.genfromtxt("msd_e_10.txt", delimiter=" ")
m3 = np.genfromtxt("msd_e_30.txt", delimiter=" ")


t = np.linspace(5e-2, .5, 100)
t1 = np.linspace(20, 100, 100)


def line(x, a, b):
    return b*x**a


f = plt.figure()

plt.plot(m1[1:, 0], m1[1:, 1], label="Solid")
plt.plot(m2[1:, 0], m2[1:, 1], label="Liquid")
plt.plot(m3[1:8000, 0], m3[1:8000, 1], label="Gas")
plt.annotate(r"$\sim t^2$", xy=[.07, .02], xytext=[.07, .02])
plt.plot(t, line(t, 2, .4), "k--")
plt.annotate(r"$\sim t^1$", xy=[20, 6], xytext=[20, 6])
plt.plot(t1, line(t1, 1, .1), "k-.")

plt.axis([1e-2, 2e2, 1e-6, 1e2])

plt.xscale("log")
plt.yscale("log")
plt.xlabel(r"$t$")
plt.ylabel("MSD(t)")
plt.legend()
plt.grid()

plt.savefig("MSD.pdf", dpi=200)

```
